#Importamos tensor y numpy, tensor 2x ya viene incluido keras

```

import numpy as np
%tensorflow_version 2.x   # Para garantizar que la versi贸n 2.x sea importada
import tensorflow as tf
print('Versi贸n de TensorFlow: ' + tf.__version__)

```

#Definimos la red Neuronal


```
model=tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.summary()
```

#Compilamos la Red, asignando la funcion de perdida y el optimizador

```
model.compile(optimizer="sgd", loss= "mean_squared_error", metrics=['accuracy'])
```

#Definimos las entradas y salidas (features y label)

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype= float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype= float)
```

#Entrenamos el modelo

```
model.fit(xs, ys, epochs=500)
```

# A la red entrenada le demos un valor que no ha visto antes

```
print(model.predict([10.0]))
```
